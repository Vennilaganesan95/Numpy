```python
import numpy as np

arr1 = np.random.randint(1, 101, 15)
print("Array:", arr1)

# Boolean masking: select values > 50
greater_than_50 = arr1[arr1 > 50]
print("Values > 50:", greater_than_50)

arr2 = np.arange(2, 41, 2)  # 20 even numbers (2 to 40)
print("Array:", arr2)

# Fancy indexing: select every third element
every_third = arr2[::3]
print("Every third element:", every_third)

arr3 = np.arange(1, 17).reshape(4, 4)
print("Array:\n", arr3)

# Boolean masking: elements > 10
greater_than_10 = arr3[arr3 > 10]
print("Elements > 10:", greater_than_10)

arr4 = np.random.randint(0, 10, 10)
print("Original Array:", arr4)

arr4[arr4 < 5] = 0
print("After replacement:", arr4)

arr5 = np.arange(1, 37).reshape(6, 6)
print("Array:\n", arr5)

selected_rows = arr5[[0, 2, 4], :]  # 0-based indexing
print("Rows 1, 3, 5:\n", selected_rows)

arr6 = np.random.randint(1, 101, 10)
print("Array:", arr6)

labels = np.where(arr6 > 50, "High", "Low")
print("Labels:", labels)

arr7 = np.random.randint(10, 51, (3, 3))
print("Array:\n", arr7)

max_index = np.unravel_index(np.argmax(arr7), arr7.shape)
print("Index of maximum value:", max_index)


arr8 = np.random.randint(0, 30, 12)
print("Array:", arr8)

conditions = [arr8 < 10, (arr8 >= 10) & (arr8 <= 20), arr8 > 20]
choices = ["Small", "Medium", "Large"]

# Specify dtype=object to allow string output
size_labels = np.select(conditions, choices, default="Unknown")
print("Size Labels:", size_labels)


arr9 = np.random.randint(-10, 10, 15)
print("Array:", arr9)

has_negative = np.any(arr9 < 0)
print("Contains negative values?", has_negative)

arr10 = np.random.randint(0, 101, 10)
print("Original Array:", arr10)

sorted_desc = np.sort(arr10)[::-1]
print("Sorted descending:", sorted_desc)

arr11 = np.arange(1, 26).reshape(5, 5)
print("Array:\n", arr11)

subset = arr11[:3, 1:4]  # rows 0-2, columns 1-3
print("Subset:\n", subset)


arr12 = np.random.rand(10)
sorted_arr12 = np.sort(arr12)
print("Sorted Array:", sorted_arr12)

value = 0.5
index = np.searchsorted(sorted_arr12, value)
print(f"Index to insert {value}:", index)


arr13 = np.random.randint(0, 10, 8)
print("Array:", arr13)

unique_sorted = np.sort(np.unique(arr13))
print("Unique sorted values:", unique_sorted)










```
