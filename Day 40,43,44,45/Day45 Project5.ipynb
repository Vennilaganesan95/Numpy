```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate dataset (1000 values from standard normal distribution)
data = np.random.randn(1000)

# Step 2: Calculate statistical metrics
mean = np.mean(data)
std_dev = np.std(data)
variance = np.var(data)

percentiles = np.percentile(data, [25, 50, 75])

print("Statistical Metrics:")
print(f"Mean: {mean:.4f}")
print(f"Standard Deviation: {std_dev:.4f}")
print(f"Variance: {variance:.4f}")
print(f"25th Percentile: {percentiles[0]:.4f}")
print(f"50th Percentile (Median): {percentiles[1]:.4f}")
print(f"75th Percentile: {percentiles[2]:.4f}")

# Step 3: Plot histogram
counts, bin_edges = np.histogram(data, bins=30)  # 30 bins for histogram
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Data (Standard Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 4: Random selection of 100 values
sample = np.random.choice(data, 100, replace=False)
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print("\nSample Analysis (100 random values):")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
```
