```python
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Generate Random Data

data = np.random.randn(1000)  # 1000 random numbers ~ Normal distribution (mean≈0, std≈1)

# Step 3: Basic Statistical Analysis

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
percentiles = np.percentile(data, [25, 50, 75])

print("Step 3: Statistical Analysis")
print(f"Mean: {mean:.4f}")
print(f"Median: {median:.4f}")
print(f"Standard Deviation: {std_dev:.4f}")
print(f"25th Percentile: {percentiles[0]:.4f}")
print(f"50th Percentile (Median): {percentiles[1]:.4f}")
print(f"75th Percentile: {percentiles[2]:.4f}")

# Step 4: Create and Visualize a Histogram

plt.hist(data, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Histogram of Random Data (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Step 5: Simulate Random Events

# Example: rolling a dice (outcomes 1–6)
outcomes = [1, 2, 3, 4, 5, 6]
random_event = np.random.choice(outcomes, size=10)  # simulate 10 dice rolls

print("\nStep 5: Random Event Simulation")
print("Simulated dice rolls:", random_event)
```
